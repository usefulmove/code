```python
homes['income_category'] = pd.cut(homes['median_income'], bins = [0, 1.5, 3, 4.5, 6, np.inf], labels = [1, 2, 3, 4, 5])
```

```python
homes
```

```python
# stratified sampling based on income category
split = StratifiedShuffleSplit(n_splits = 1, test_size = 0.2, random_state = 618)
```

```python
for train_index, test_index in split.split(homes, homes.income_category):
    strat_train_set = homes.loc[train_index]
    strat_test_set = homes.loc[test_index]
```

```python
strat_train_set
```

```python
strat_test_set
```

```python
for dframe in (strat_test_set, strat_train_set):
    dframe.drop(columns = 'income_category', inplace = True)
```
