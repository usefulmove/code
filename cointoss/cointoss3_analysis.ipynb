```python
from plotnine import ggplot, aes, geom_col, geom_text, labs, coord_flip
from plotnine.themes import theme_set, theme_minimal

theme_set(theme_minimal())
n_ptrns = len(df)

(ggplot(df[-(n_ptrns >> 1):], aes(x='pattern', y='tosses_to_match'))
    + geom_col(fill='#87ffaf', color='#000', alpha=0.6)
    + geom_text(aes(label='round(tosses_to_match, 1)'), nudge_y=-2.5, size=8)
    + labs(title='coin flips req. to match a given pattern', x='', y='avg. no. coin flips to match pattern')
    + coord_flip()
)
```
