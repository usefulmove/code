## import data

```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
nba = pd.read_csv('/content/drive/MyDrive/data/nba.csv', sep=';', encoding='latin1')
pd.set_option('display.max_columns', 100)
display(nba[:8])
```

## clean dataset

```python
print(nba.columns)
```

```python
# calculate normalized shot efficiency (points per shot)
nba['pps'] = (nba['3P'] * 3 + nba['2P'] * 2 + nba['FT']) / (nba['3PA'] + nba['2PA'] + (nba['FTA'] / 2))
```

## analysis

```python
# 3 point percentage
(nba
    .sort_values('3P%', ascending=False)
[nba['3PA'] > 4] # more than four 3 point attempts per game
    [nba['G'] > 20] # more than 20 games played
    [nba['PTS'] > 12] # more than 12 points per game
    [:20]
)
```

```python
# points per shot attempt (pps) - non-centers
pps = (nba
    .sort_values('pps', ascending=False) # sort by overall efficiency
    [nba['Pos'] != 'C']
    [nba['G'] > 20] # more than 20 games played
    [nba['PTS'] > 12] # more than 12 points per game
    [nba['MP'] > 24] # more than 24 minutes per game
)
pps = pps.drop(174) # remove BRK Kevin Durant (total for the year is included)
pps = pps.reset_index(drop=True)
display(pps[:30])
```

```python
# points per shot attempt (pps) - non-centers
pps = (nba
    .sort_values('pps', ascending=False) # sort by overall efficiency
    [nba['Pos'] != 'C']
    [nba['G'] > 20] # more than 20 games played
    [nba['PTS'] > 12] # more than 12 points per game
    [nba['MP'] > 24] # more than 24 minutes per game
)
pps = pps.drop(174) # remove BRK Kevin Durant (total for the year is included)
pps = pps.reset_index(drop=True)
display(pps[:30])
```

```python
# free throws per shot attempt
nba['fta_fga'] = nba['FTA'] / nba['FGA']
(nba
    .sort_values('fta_fga', ascending=False)
    [nba['PTS'] > 12] # more than 12 points per game
    [nba['MP'] > 24] # more than 24 minutes per game
    [nba['FGA'] > 15] # more than 15 shot attempts per game
    [:12]
)
```

## playoffs

```python
# playoffs
poffs = pd.read_csv('/content/drive/MyDrive/data/nba_playoffs.csv', sep=';', encoding='latin1')
poffs['pps'] = (poffs['3P'] * 3 + poffs['2P'] * 2 + poffs['FT']) / (poffs['3PA'] + poffs['2PA'] + (poffs['FTA'] / 2))
#display(poffs[:8])

ppps = (poffs
    .sort_values('pps', ascending=False) # sort by overall efficiency
    [poffs['PTS'] > 12] # more than 12 points per game
    [poffs['MP'] > 24] # more than 24 minutes per game
    [poffs['FGA'] > 15] # more than 15 shot attempts per game
)
ppps = ppps.reset_index(drop=True)
display(ppps[:15])
```
