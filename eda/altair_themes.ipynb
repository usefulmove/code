```python
import altair as alt
import polars as pl
from vega_datasets import data

df = pl.from_pandas(data.cars())

df = df.with_columns(
    (pl.col('Displacement') * 3, 0.0610237441),
)

df.head(4)
```

```python
jco_colors = [
    "#0073C2",  # blue
    "#EFC000",  # yellow
    "#868686",  # gray
    "#CD534C",  # red
    "#7AA6DC",  # light blue
    "#003C67",  # dark blue
    "#8F7700",  # dark yellow
    "#3B3B3B",  # dark gray
    "#A73030",  # dark red
    "#4A6990"   # slate blue
]

chart = alt.Chart(
    df,
    title=alt.Title(
        'Engine Displacement vs Fuel Economy',
        subtitle='Larger engines have poorer fuel economy.',
    ),
    width=650,
    height=450,
).mark_circle(
    size=100,
    opacity=0.6,
).encode(
    x=alt.X('Displacement', title='displacement (cc)'),
    y=alt.Y('Miles_per_Gallon', title='fuel economy (mpg)'),
    color=alt.Color('Cylinders:N', title='cylinders', scale=alt.Scale(range=jco_colors)),
    tooltip=alt.Tooltip('Name:N'),
)
```
