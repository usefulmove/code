```python
file = '/content/drive/MyDrive/data/For Public_ Whose Heritage Master Sheet - Whose Heritage 2019 7.28.19 Master .tsv'
df = pd.read_csv(file, sep='\t')
df
```

```python
# data conversions
df['Honoree'] = df['Honoree'].astype('category')
df['City'] = df['City'].astype('category')
df['County'] = df['County'].astype('category')
df['State'] = df['State'].astype('category')
df['Year Dedicated'] = pd.to_numeric(df['Year Dedicated'], errors='coerce').astype('Int64')
df['Year Removed'] = pd.to_numeric(df['Year Removed'], errors='coerce').astype('Int64')
df['Was this symbol removed?'] = df['Was this symbol removed?'].astype('category')
df[['Latitude', 'Longitude']] = df['Coordinates'].str.split(',', expand=True)
df['Latitude'] = pd.to_numeric(df['Latitude'], errors='coerce')
df['Longitude'] = pd.to_numeric(df['Longitude'], errors='coerce')
```

```python
from plotnine import (
    ggplot,
    aes,
    geom_bar,
    geom_text,
    geom_segment,
    labs,
    theme,
    theme_minimal,
    element_text,
    element_line,
    scale_y_continuous,
)

bbc_style_theme = (
    theme_minimal(base_family="DejaVu Sans", base_size=12)
    + theme(
        panel_grid_major=element_line(color="#e0e0e0", size=0.5),
        panel_grid_minor=element_line(color="#f5f5f5", size=0.25),
        axis_title=element_text(size=11, color="#333333"),
        axis_text=element_text(size=10, color="#333333"),
        plot_title=element_text(size=14, weight="normal", color="#222222"),
        plot_subtitle=element_text(size=12, color="#444444"),
        legend_position="none",
    )
)

df_year_clean = df.dropna(subset=['Year Dedicated']).astype({'Year Dedicated': 'int64'})

(
    ggplot(df_year_clean, aes(x='Year Dedicated'))
    + geom_bar(fill='red', color='black', alpha=0.618)
    # Civil War
    + geom_text(
        x=1865+1, y=45, label='Civil War ends',
        color='#980606',
        ha='left', va='center',
        size=10, angle=0,
        inherit_aes=False
    )
    + geom_segment(
        x=1865, xend=1865,
        y=0, yend=45-1,
        size=0.2,
        color='grey'
    )
    # Plessy v. Ferguson
    + geom_text(
        x=1896-1, y=35, label='Plessy v. Ferguson\n("separate but equal")',
        color='#980606',
        ha='right', va='center',
        size=10, angle=0,
        inherit_aes=False
    )
    + geom_segment(
        x=1896, xend=1896,
        y=0, yend=35-1,
        size=0.2,
        color='grey'
    )
    # The Clansman
    + geom_text(
        x=1905+1, y=55, label='The Clansman: A Historical\nRomance of the Ku Klux Klan released',
        color='#980606',
        ha='left', va='center',
        size=10, angle=0,
        inherit_aes=False
    )
    + geom_segment(
        x=1905, xend=1905,
        y=0, yend=55-1,
        size=0.2,
        color='grey'
    )
    # World War II
    + geom_text(
        x=1945+1, y=50, label='World War II ends',
        color='#980606',
        ha='left', va='center',
        size=10, angle=0,
        inherit_aes=False
    )
    + geom_segment(
        x=1945, xend=1945,
        y=0, yend=50-1,
        size=0.2,
        color='grey'
    )
    # Civil rights movement starts
    + geom_text(
        x=1954+1, y=45, label='Civil rights movement starts',
        color='#980606',
        ha='left', va='center',
        size=10, angle=0,
        inherit_aes=False
    )
    + geom_segment(
        x=1954, xend=1954,
        y=0, yend=45-1,
        size=0.2,
        color='grey'
    )
    # Civil Rights Act of 1964
    + geom_text(
        x=1964+1, y=40, label='Civil Rights Act of 1964',
        color='#980606',
        ha='left', va='center',
        size=10, angle=0,
        inherit_aes=False
    )
    + geom_segment(
        x=1964, xend=1964,
        y=0, yend=40-1,
        size=0.2,
        color='grey'
    )
    + scale_y_continuous(limits=(0, 56))
    + labs(
        title='U.S. Confederate Monument Dedications by Year',
        x='',
        y='no. of publicly-supported symbols'
    )
    + bbc_style_theme
    + theme(figure_size=(17, 6))
)
```
